```python
import numpy as np
import pandas as pd
```

```python
#create a runnning attendance % for the students in the attendence cvs
# Present = 1, Absent = 0, Half-day = 0.5, Tardy = 0.9
#read cvs files in for attendance

df = pd.read_csv('untidy-data/attendance.csv')
```

```python
df
```

```python
#rename unnamed column 
df = df.rename(columns={'Unnamed: 0': 'name'})
df
```

```python
#melt data so it can be aggregated better
df = pd.melt(df, id_vars='name', var_name= 'date', value_name='att')
```

```python
#create a new variable that holds the value of each letter
```

```python
df.loc[df['att'] == 'P', 'percent'] = 1
df.loc[df['att'] == 'H', 'percent'] = 0.5
df.loc[df['att'] == 'A', 'percent'] = 0
df.loc[df['att'] == 'T', 'percent'] = .9
```

```python
#how many tardies, if 10, add on absance 
df.where(df.att == 'T').groupby('name').att.count()
```

```python
#group by name to getting the avg % of attendence  for each student
df.groupby('name').percent.mean()
```

```python
#load csv files into data frame
cake_df = pd.read_csv('untidy-data/cake_recipes.csv')
```

```python
#split column name for recipe into 2 parts, use the first part
rename = cake_df.columns[0].split(':')
```

```python
cake_df = cake_df.rename(columns={'recipe:position': rename[0]})
```

```python
cake_df
```

```python
##condense data into a better form for aggregation
cake_df = pd.melt(cake_df, id_vars= rename[0], var_name= "temp", value_name='value')
```

```python
#split the rack position and the recipe type into its own variables than remove redundent column
cake_df[['recipe_type', 'rack']] = cake_df.recipe.str.split(':', expand=True)
```

```python
cake_df = cake_df.drop(columns={'recipe'})
```

```python
cake_df
```

```python
#a. highest avg for recipe type = b
cake_df.groupby('recipe_type').value.mean()
```

```python
#b. highest avg for temp = 275
cake_df.groupby('temp').value.mean()
```

```python
#c highest value for temp,recipe type and rack position

#finds the highest value of the value column givin the conditions to group by
cake_max = cake_df.groupby(['recipe_type','temp', 'rack']).value.mean().reset_index().value.max()
cake_max
```

```python
#returns the row where the value is equal to the highest avg
cake_df.loc[cake_df.value == cake_max]
```
